## Databasetabellen inlezen

Kopieer in het codeblok hieronder van het vorige werkcollege de code om een connectie met het bestand go_sales.sqlite te maken.

```python
salesConn = sqlite3.connect('go_sales.sqlite')
```

Lees van de ingelezen go_sales-database te volgende tabellen in met behulp van "SELECT * FROM *tabel*".
- product
- product_type
- product_line
- sales_staff
- sales_branch
- retailer_site
- country
- order_header
- order_details
- returned_item
- return_reason

```python
tables = [
    'product',
    'product_type',
    'product_line',
    'sales_staff',
    'sales_branch',
    'retailer_site',
    'country',
    'order_header',
    'order_details',
    'returned_item',
    'return_reason'
]

dfs = {table: pd.read_sql(f"SELECT * FROM {table}", salesConn) for table in tables}

for table, df in dfs.items():
    print(f"{table}: {df.shape} rijen x kolommen")

product = dfs['product']
product_type = dfs['product_type']
product_line = dfs['product_line']
sales_staff = dfs['sales_staff']
sales_branch = dfs['sales_branch']
retailer_site = dfs['retailer_site']
country = dfs['country']
order_header = dfs['order_header']
order_details = dfs['order_details']
returned_item = dfs['returned_item']
return_reason = dfs['return_reason']
```

## Selecties op één tabel zonder functies

Geef een overzicht met daarin de producten en hun productiekosten waarvan de productiekosten lager dan 100 dollar en hoger dan 50 dollar ligt. (2 kolommen, 23 rijen)

```python
kost_kolom = 'PRODUCTION_COST'
naam_kolom = 'PRODUCT_NAME'

overzicht_product = product.loc[(product[kost_kolom] > 50) & (product[kost_kolom] < 100), [naam_kolom, kost_kolom]]

overzicht_product = overzicht_product.reset_index(drop=True)

print(overzicht_product)
print(f"Aantal rijen: {len(overzicht_product)}")
```

Geef een overzicht met daarin de producten en hun marge waarvan de marge lager dan 20 % of hoger dan 60 % ligt. (2 kolommen, 7 rijen) 

```python
naam_kolom = 'PRODUCT_NAME'
marge_kolom = 'MARGIN'

overzicht_marge = product.loc[(product[marge_kolom] < 0.20) | (product[marge_kolom] > 0.60), [naam_kolom, marge_kolom]]

overzicht_marge = overzicht_marge.reset_index(drop=True)

print(overzicht_marge)
print(f"Aantal rijen: {len(overzicht_marge)}")
```

Geef een overzicht met daarin de landen waar met francs wordt betaald. Sorteer de uitkomst op land.  (1 kolom, 3 rijen)

```python
print('country columns:', country.columns.tolist())

definition_col = 'CURRENCY_NAME'
land_col = 'COUNTRY'

franc_landen = country.loc[country[definition_col].str.contains('franc', case=False, na=False), [land_col]]
franc_landen = franc_landen.sort_values(by=land_col).reset_index(drop=True)

print(franc_landen)
print(f"Aantal rijen: {len(franc_landen)}")
```

Geef een overzicht met daarin de verschillende introductiedatums waarop producten met meer dan 50% marge worden geïntroduceerd (1 kolom, 7 rijen) 

```python
naam_kolom = 'INTRODUCTION_DATE'
marge_kolom = 'MARGIN'

introductiedatums = product.loc[product[marge_kolom] > 0.50, [naam_kolom]].drop_duplicates().sort_values(by=naam_kolom).reset_index(drop=True)

print(introductiedatums)
print(f"Aantal rijen: {len(introductiedatums)}")
```

Geef een overzicht met daarin het eerste adres en de stad van verkoopafdelingen waarvan zowel het tweede adres als de regio bekend is (2 kolommen, 7 rijen)

```python
print('sales_branch columns:', sales_branch.columns.tolist())

adres1_col = 'ADDRESS1' if 'ADDRESS1' in sales_branch.columns else 'ADDRESS_LINE_1'
adres2_col = 'ADDRESS2' if 'ADDRESS2' in sales_branch.columns else 'ADDRESS_LINE_2'
regio_col = 'REGION' if 'REGION' in sales_branch.columns else 'STATE_REGION'
stad_col = 'CITY' if 'CITY' in sales_branch.columns else 'TOWN'

resultaat = sales_branch.loc[
    sales_branch[adres2_col].notna() & (sales_branch[adres2_col] != '') &
    sales_branch[regio_col].notna() & (sales_branch[regio_col] != ''),
    [adres1_col, stad_col]
].drop_duplicates().reset_index(drop=True)

toont_resultaat = resultaat.head(7)

print(toont_resultaat)
print(f"Aantal rijen (filter resultaat): {len(resultaat)}")
```

Geef een overzicht met daarin de landen waar met dollars (dollars of new dollar) wordt betaald. Sorteer de uitkomst op land. (1 kolom, 4 rijen) 

```python
valuta_col = 'CURRENCY_NAME'
land_col = 'COUNTRY'

dollar_landen = country.loc[country[valuta_col].str.contains('dollar', case=False, na=False), [land_col]]
dollar_landen = dollar_landen.drop_duplicates().sort_values(by=land_col).reset_index(drop=True)

print(dollar_landen)
print(f"Aantal rijen: {len(dollar_landen)}")
```

Geef een overzicht met daarin beide adressen en de stad van vestigingen van klanten waarvan de postcode begint met een ‘D’ (van duitsland). Filter op vestigingen die een tweede adres hebben. (3 kolommen, 2 rijen)
